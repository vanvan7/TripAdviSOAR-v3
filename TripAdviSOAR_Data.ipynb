```python
import pandas as pd
import numpy as np
```

```python
t_soar_db = pd.read_csv('tripadvisor_european_restaurants.csv', nrows=1000)
t_soar_db.head()
```

```python
clean_t_soar_db = t_soar_db[['restaurant_name', 'address', 'price_level', 'cuisines', 'special_diets' ]]
clean_t_soar_db.head()
```

```python
username = clean_t_soar_db['restaurant_name']+"1"
clean_t_soar_db.insert(loc=0, column='username', value=username)
clean_t_soar_db['username'] = clean_t_soar_db['username'].str.lower()
clean_t_soar_db['username'] = clean_t_soar_db['username'].str.replace(" ", "")
clean_t_soar_db
```

```python
email = clean_t_soar_db['username']+"@gmail.com"
clean_t_soar_db.insert(loc=2, column='email', value=email)

clean_t_soar_db
```

```python
! pip install names
```

```python
import names
restaurant_owner = []

for i in range(len(clean_t_soar_db['restaurant_name'])):
  restaurant_owner.append(names.get_full_name())

clean_t_soar_db.insert(loc=3, column='restaurant_owner', value=restaurant_owner)

clean_t_soar_db.head()
```

```python
clean_t_soar_db.insert(loc=5, column='opening_hours', value='17:00-23:00')
```

```python
clean_t_soar_db = clean_t_soar_db.rename(columns={'price_level': 'price', 'cuisines':'cooking_type', 'special_diets':'specialdiet'})
clean_t_soar_db
```

```python
import random
contact_list=[]
for i in range(len(clean_t_soar_db)):
  phone_number='07'
  for i in range(1,8):
    phone_number = phone_number + str(random.randint(0,9))
  contact_list.append(phone_number)
clean_t_soar_db.insert(loc=8, column='contact', value=contact_list)
```

```python
from faker import Faker
from faker_food import FoodProvider

fake = Faker()
fake.add_provider(FoodProvider)

menu_list=[]
for i in range(len(clean_t_soar_db)):
  menu = ''
  for i in range(1,5):
    menu = menu + str(fake.dish()) + ', '
  menu_list.append(menu)
clean_t_soar_db.insert(loc=9, column='menu', value=menu_list)
```
