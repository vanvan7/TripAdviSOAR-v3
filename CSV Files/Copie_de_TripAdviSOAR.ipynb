```python
import pandas as pd
import numpy as np
```

```python
t_soar_db = pd.read_csv('tripadvisor_european_restaurants.csv', nrows=1000)
t_soar_db.head()
```

```python
clean_t_soar_db = t_soar_db[['restaurant_name', 'address', 'price_level', 'cuisines', 'special_diets' ]]
clean_t_soar_db.head()
```

```python
username = clean_t_soar_db['restaurant_name']+"1"
clean_t_soar_db.insert(loc=0, column='username', value=username)
clean_t_soar_db['username'] = clean_t_soar_db['username'].str.lower()
clean_t_soar_db['username'] = clean_t_soar_db['username'].str.replace(" ", "")
clean_t_soar_db
```

```python
from faker import Faker
from faker_food import FoodProvider

fake = Faker()
fake.add_provider(FoodProvider)

menu_list=[]
for i in range(len(clean_t_soar_db)):
  menu = ''
  for i in range(1,5):
    menu = menu + str(fake.dish()) + ', '
  menu_list.append(menu)
clean_t_soar_db.insert(loc=9, column='menu', value=menu_list)
```

```python
def java_string_hashcode(s):
    """Mimic Java's hashCode in python 2"""
    h = 0
    for c in s:
        h = int((((31 * h + ord(c)) ^ 0x80000000) & 0xFFFFFFFF) - 0x80000000)
    return h
```

```python
password_list = []
hashed_password_list = []
for i in range(len(clean_t_soar_db)):
  password=''
  for i in range(1,8):
    password = password + str(random.randint(0,9))

  hashed_password = java_string_hashcode(password)  
  password_list.append(password)
  hashed_password_list.append(hashed_password)

password_t_soar_db = clean_t_soar_db.copy(deep=True)
password_t_soar_db.insert(loc=11, column='password', value=password_list)
password_t_soar_db.to_csv('Passwords.csv', index=False, encoding = 'utf-8-sig')
#password_t_soar_db.to_csv('/content/drive/MyDrive/Colab Notebooks/Passwords.csv', index=False, encoding = 'utf-8-sig')

clean_t_soar_db.insert(loc=11, column='password', value=hashed_password_list)
```
